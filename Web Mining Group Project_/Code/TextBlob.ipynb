```python
import pandas as pd 
import numpy as np
from nltk.tokenize import RegexpTokenizer
from textblob import TextBlob
```

```python
pd.read_csv('new_Brooklyn.csv')
```

```python
## Data preprocess

df1 = pd.read_csv('new_Brooklyn.csv')
df1 = pd.DataFrame(df1, columns=['review_count', 'rating','price', 'reviews', 'is_closed'])
df1 = df1.dropna()
p={'$$':2, '$':1, '$$$':3, '$$$$':4}
df1['price']=df1['price'].map(p)


df1['is_closed'] = df1['is_closed']+0
df1.index = range(len(df1))
```

```python
## Data preprocess

df2 = pd.read_csv('new_Manhattan.csv')
df2 = pd.DataFrame(df2, columns=['review_count', 'rating','price', 'reviews', 'is_closed'])
df2 = df2.dropna()
p={'$$':2, '$':1, '$$$':3, '$$$$':4}
df2['price']=df2['price'].map(p)


df2['is_closed'] = df2['is_closed']+0
df2.index = range(len(df2))
```

```python
df2
```

```python
df1['polarity'] = [TextBlob(review).polarity for review in df1['reviews']]
df1['subjectivity'] = [TextBlob(review).subjectivity for review in df1['reviews']]
df1 = df1.drop(columns=['reviews'])

df2['polarity'] = [TextBlob(review).polarity for review in df2['reviews']]
df2['subjectivity'] = [TextBlob(review).subjectivity for review in df2['reviews']]
df2 = df2.drop(columns=['reviews'])
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
train_X = np.array(df2.drop(columns=['is_closed']))
train_y = np.array(df2['is_closed'])

test_X = np.array(df1.drop(columns=['is_closed']))
test_y = np.array(df1['is_closed'])

rfc = RandomForestClassifier(n_estimators=100, random_state=42)
rfc.fit(train_X, train_y)
rfc_test_score = rfc.score(test_X, test_y)
print('Test Accuracy:', rfc_test_score)
```

```python
rfc = SVC()
rfc.fit(train_X, train_y)
rfc_test_score = rfc.score(test_X, test_y)
print('Test Accuracy:', rfc_test_score)
```

```python
rfc = LogisticRegression()
rfc.fit(train_X, train_y)
rfc_test_score = rfc.score(test_X, test_y)
print('Test Accuracy:', rfc_test_score)
```

```python
rfc = LinearSVC()
rfc.fit(train_X, train_y)
rfc_test_score = rfc.score(test_X, test_y)
print('Test Accuracy:', rfc_test_score)
```

```python
rfc = SGDClassifier()
rfc.fit(train_X, train_y)
rfc_test_score = rfc.score(test_X, test_y)
print('Test Accuracy:', rfc_test_score)
```

```python
rfc = GradientBoostingClassifier()
rfc.fit(train_X, train_y)
rfc_test_score = rfc.score(test_X, test_y)
print('Test Accuracy:', rfc_test_score)
```
